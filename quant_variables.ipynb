Visualizing Two Quantitative Variables

- Introduction to relplot()

Use relplot() to create relational plots with scatter or line plots. It lets you create subplots in a single figure.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
# Default Relational Plot
sns.relplot(x="total_bill", y="tip", data=tips, kind="scatter")
plt.show()
```

```python
# Relational Plot Separating Plots by Smokers and Non-Smokers in Columns
sns.relplot(x="total_bill", y="tip", data=tips, kind="scatter", col="smoker")
plt.show()
```

```python
# Relational Plot Separating Plots by Smokers and Non-Smokers in Rows
sns.relplot(x="total_bill", y="tip", data=tips, kind="scatter", row="smoker")
plt.show()
```

```python
# Relational Plot Separating Plots by Smokers and Non-Smokers and Time of the Day
sns.relplot(x="total_bill", y="tip", data=tips, kind="scatter", col="smoker", row="time")
plt.show()
```

```python
# Relational Plot Separating Plots by Days of the Week and Setting two plots per row
sns.relplot(x="total_bill", y="tip", data=tips, kind="scatter", col="day", col_wrap=2)
plt.show()
```

```python
# Relational Plot Separating Plots by Days of the Week and Setting two plots per row. 
# Also defining the order of the days of the week we want by setting the col_order parameter.
sns.relplot(x="total_bill", y="tip", data=tips, kind="scatter", col="day", col_wrap=2, col_order = ["Sat","Fri", "Sun", "Thur"])
plt.show()
```

1. Creating subplots with col and row

```python
student_data = pd.read_csv('student-alcohol-consumption.csv')
student_data.head()
```

```python
# Change to use relplot() instead of scatterplot()
sns.relplot(x="absences", y="G3", 
                data=student_data, kind="scatter")

# Show plot
plt.show()
```

```python
# Change to make subplots based on study time
sns.relplot(x="absences", y="G3", 
            data=student_data,
            kind="scatter", col="study_time")

# Show plot
plt.show()
```

```python
# Change this scatter plot to arrange the plots in rows instead of columns
sns.relplot(x="absences", y="G3", 
            data=student_data,
            kind="scatter", 
            row="study_time")

# Show plot
plt.show()
```

2. Creating two-factor subplots

Here, we want to answer the following question: does a student's first semester grade ("G1") tend to correlate with their final grade ("G3")?

There are many aspects of a student's life that could result in a higher or lower final grade in the class. For example, some students receive extra educational support from their school ("schoolsup") or from their family ("famsup"), which could result in higher grades. 

```python
# Create a scatter plot of G1 vs. G3
sns.relplot(x="G1", y="G3", data=student_data, kind="scatter")

# Show plot
plt.show()
```

```python
# Adjust to add subplots based on school support
sns.relplot(x="G1", y="G3", 
            data=student_data,
            kind="scatter", col="schoolsup", col_order=["yes","no"])

# Show plot
plt.show()
```

```python
# Adjust further to add subplots based on family support
sns.relplot(x="G1", y="G3", 
            data=student_data,
            kind="scatter", 
            col="schoolsup",
            row="famsup",
            row_order=["yes","no"], col_order=["yes","no"])

# Show plot
plt.show()
```

 Conclusion: It looks like the first semester grade does correlate with the final grade, regardless of what kind of support the student received.

- Customizing scatter plots

Seaborn allows you to customize the size, style and transparency of the points.

1. Point Size - to vary the size of a point based on a variable

```python
# Here we add size to vary the size of a point based on it
sns.relplot(x="total_bill", y="tip", data=tips, kind="scatter", size="size")
plt.show()
```

```python
# Here we use size and hue parameters combined to make the graph easier to read.
sns.relplot(x="total_bill", y="tip", data=tips, kind="scatter", size="size", hue="size")
plt.show()
```

2. Point Style - to use different point styles for each value of the variable

```python
# Hue creates different colored points based on smoking status and style changes their style.
sns.relplot(x="total_bill", y="tip", data=tips, kind="scatter", hue="smoker", style="smoker")
plt.show()
```

3. Point Transparency - to vary the transparency of the point

```python
# Set the variable alpha to be between 0 and 1
sns.relplot(x="total_bill", y="tip", data=tips, kind="scatter", alpha=0.4)
plt.show()
```

- Line Plots

1. Interpreting line plots

In this exercise, we'll continue to explore Seaborn's mpg dataset, which contains one row per car model and includes information such as the year the car was made, its fuel efficiency (measured in "miles per gallon" or "M.P.G"), and its country of origin (USA, Europe, or Japan).

How has the average miles per gallon achieved by these cars changed over time?

```python
# Create line plot
sns.relplot(x="model_year", y="mpg", kind="line", data=mpg)

# Show plot
plt.show()
```

The shaded region represents a confidence interval for the mean, not the distribution of the observations.

2. Visualizing standard deviation with line plots

```python
# Make the shaded area show the standard deviation
sns.relplot(x="model_year", y="mpg",
            data=mpg, kind="line", ci="sd")

# Show plot
plt.show()
```

This plot shows us the distribution of miles per gallon for all the cars in each year.

3. Plotting subgroups in line plots

```python
# Create line plot of model year vs. horsepower
# ci parameter is used to turn off the confidence intervals
sns.relplot(x="model_year", y="horsepower", 
            data=mpg, kind="line", 
            ci=None)

# Show plot
plt.show()
```

```python
# Change to create subgroups for country of origin
sns.relplot(x="model_year", y="horsepower", 
            data=mpg, kind="line", 
            ci=None, style="origin", hue="origin")

# Show plot
plt.show()
```

```python
# Add markers and make each line have the same style
# Use the dashes parameter to use solid lines for all countries, 
# while still allowing for different marker styles for each line.
sns.relplot(x="model_year", y="horsepower", 
            data=mpg, kind="line", 
            ci=None, style="origin", 
            hue="origin", markers=True,
            dashes=False)

# Show plot
plt.show()
```

Now that we've added subgroups, we can see that this downward trend in horsepower was more pronounced among cars from the USA.
