Customizing Plots

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

If we only care about the comparisons between groups or the general trend, this is a good choice of graph:

```python
student_data = pd.read_csv('student-alcohol-consumption.csv')
student_data.head()

# Create a point plot that uses color to create subgroups
sns.catplot(x="romantic", y="absences", data=student_data, kind="point", hue="school")

# Show plot
plt.show()
```

WHITEGRID: Changing the style to "whitegrid" will add a gray grid in the background. This is interesting if you want the audience to determine specific values of the plotted points instead of making higher level observations.

```python
# Set style
sns.set_style("whitegrid")

# Create a point plot that uses color to create subgroups
sns.catplot(x="romantic", y="absences", data=student_data, kind="point", hue="school")

# Show plot
plt.show()
```

TICKS: This adds small tick marks to the x and y axes.

```python
# Set style
sns.set_style("ticks")

# Create a point plot that uses color to create subgroups
sns.catplot(x="romantic", y="absences", data=student_data, kind="point", hue="school")

# Show plot
plt.show()
```

DARK: Provides a gray background.

```python
# Set style
sns.set_style("dark")

# Create a point plot that uses color to create subgroups
sns.catplot(x="romantic", y="absences", data=student_data, kind="point", hue="school")

# Show plot
plt.show()
```

DARKGRID: Provides a gray background with a white grid.

```python
# Set style
sns.set_style("darkgrid")

# Create a point plot that uses color to create subgroups
sns.catplot(x="romantic", y="absences", data=student_data, kind="point", hue="school")

# Show plot
plt.show()
```

SET PALLETE FUNCTION: It is used to change the colors of the plot. You can use the pre-defined ones or create your own palette.

**Diverging Palettes** 

"RdBu", "PRGn", "RdBu_r", "PRGn_r" - the "_r" is used to reverse the palette.

```python
survey_data = pd.read_csv('young-people-survey-responses.csv')
survey_data.head()

# Set palette
sns.set_palette("RdBu")

# Set Style
sns.set_style("white")

# List of categories from lowest to highest
category_order = ["<2 hours", 
                  "2 to 5 hours", 
                  "5 to 10 hours", 
                  ">10 hours"]

# Turn off the confidence intervals
sns.catplot(x="study_time", y="G3",
            data=student_data,
            kind="bar",
            order=category_order, ci=None)

# Show plot
plt.show()
```

**Sequential Palettes**

Single colors moving from light to dar values. They are great for emphasizing a variable on a continuous scale.

"Greys", "Blues", "PuRd", "GnBu" are a few examples.

```python
mpg = pd.read_csv('mpg.csv')
mpg.head()

# Set palette
sns.set_palette("GnBu")

# Create scatter plot of horsepower vs. mpg
sns.relplot(x = "horsepower", y="mpg", data=mpg, kind="scatter", size="cylinders")

# Show plot
plt.show()
```

**CUSTOM PALETTES**

You can create your own by passing in a list of color names. or hex color codes.

```python
# Custom palette using color names

custom_palette = ["red", "green", "orange", "blue", "yellow"]
sns.set_palette(custom_palette)
```

```python
# Custom palette using hex codes

custom_palette2 = ["#FBB4AE", "#DECBE4", "#FED9A6", "#FFFFCC", "#F2F2F2"]
sns.set_palette(custom_palette2)
```

**CHANGING SCALES**

You can change the scale of the plot by using the function "set context".

The options are: "paper", "notebook", "talk", "poster". The default one is "paper".

```python
student_data = pd.read_csv('student-alcohol-consumption.csv')
student_data.head()

# Set context
sns.set_context("talk")

# Use customized palette
custom_palette2 = ["red", "green"]
sns.set_palette(custom_palette2)

# Create a point plot that uses color to create subgroups
sns.catplot(x="romantic", y="absences", data=student_data, kind="point", hue="school")

# Show plot
plt.show()
```

**CREATING INFORMATIVE VISUALIZATIONS**

**Adding titles and labels: Part 1**

Seaborn plots create two different types of objects: FacetGrid and AxesSubplot.

To figure out what type of plot you are working with, assign the plot to a variable (often names "g") and write type(g) to return the object type.

FacetGrid: consists of one or more AxesSubplots >> relplot(), catplot().

AxesSubplot: only creates a single plot >> scatterplot(), countplot().

- Adding a title to FacetGrid and adjust the height with the y parameter

```python
# Import database
survey_data = pd.read_csv('young-people-survey-responses.csv')
survey_data.head()

# Set the style to "whitegrid"
sns.set_style("whitegrid")

# Set context
sns.set_context("paper")

# List of categories from lowest to highest
category_order = ["<2 hours", 
                  "2 to 5 hours", 
                  "5 to 10 hours", 
                  ">10 hours"]

# Turn off the confidence intervals
g = sns.catplot(x="study_time", y="G3",
            data=student_data,
            kind="bar",
            order=category_order, ci=None)

# Assign Title
g.fig.suptitle("NEW TITLE", y=1.03)

# Show plot
plt.show()
```

**Adding titles and labels: Part 2**

- Adding a title to AxesSubplot

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
# Create Plot
g = sns.boxplot(x="time", y="total_bill", data=tips)

# Set Title
g.set_title("New Title", y=1.03)

plt.show()
```

- Titles for Subplots

Use fig.suptitle to add a title for the figure as a whole:

```python
# Create Plot
g = sns.catplot(x="time", y="total_bill", data=tips, kind="box", col="sex")

# Set Title for Plot
g.fig.suptitle("New Title", y=1.03)

plt.show()
```

Use set_titles to set the titles for each AxesSubplot.

```python
# Create Plot
g = sns.catplot(x="time", y="total_bill", data=tips, kind="box", col="sex")

# Set Title for Plot
g.fig.suptitle("New Title", y=1.03)

# Set Title for SubPlots
g.set_titles("This is {col_name}")

plt.show()
```

- Adding Axis Labels : this works with the two types of objects

```python
# Create Plot
g = sns.catplot(x="time", y="total_bill", data=tips, kind="box", col="sex")

# Set Title for Plot
g.fig.suptitle("New Title", y=1.03)

# Set Title for SubPlots
g.set_titles("This is {col_name}")

# Adding Axis Labels
g.set(xlabel="New X Label",
      ylabel="New Y Label")

plt.show()
```

- Rotating Tick Labels : this works with the two types of objects

```python
# Create Plot
g = sns.catplot(x="time", y="total_bill", data=tips, kind="box")

# Set Title for Plot
g.fig.suptitle("New Title", y=1.03)

# Adding Axis Labels
g.set(xlabel="New X Label",
      ylabel="New Y Label")

# Rotate tick labels
plt.xticks(rotation = 90)

plt.show()
```
