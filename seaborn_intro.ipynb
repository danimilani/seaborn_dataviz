```python
# import database Countries of the World as Dataframe
df = pd.read_csv('countries-of-the-world.csv')
```

```python
# Show Dataframe
df.head()
```

```python
# Convert Dataframe columns to lists
gdp = df["GDP ($ per capita)"].values.tolist()
phones = df["Phones (per 1000)"].values.tolist()
percent_literate = df["Literacy (%)"].values.tolist()
```

Create scatter plot with GDP on the x-axis and number of phones on the y-axis

```python
# Create scatter plot with GDP on the x-axis and number of phones on the y-axis
sns.scatterplot(x=gdp, y=phones)
plt.show()
```

```python
# Change this scatter plot to have percent literate on the y-axis
sns.scatterplot(x=gdp, y=percent_literate)

# Show plot
plt.show()
```

Adding a third variable with hue

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
# Adding hue makes it automatically color each point by whether they are smoker or not and adds a legend to the plot.
sns.scatterplot(x="total_bill", y="tip", data=tips, hue="smoker")
plt.show()
```

```python
# The hue_order parameter takes in a list of values and will set the order of the values in the plot accordingly
sns.scatterplot(x="total_bill", y="tip", data=tips, hue="smoker", hue_order=["Yes", "No"])
plt.show()
```

```python
# Assign colors to the values according to what you want

# Create a dictionary the maps the values to the colors
hue_colors = {"Yes":"black", "No":"red"}

# Set Hue and Palette
sns.scatterplot(x="total_bill", y="tip", data=tips, hue="smoker", palette=hue_colors)
plt.show()
```

```python
# Assign colors to the values according to what you want - Using HTML HEX CODES

# Create a dictionary the maps the values to the colors
hue_colors = {"Yes":"#808080", "No":"#00FF00"}

# Set Hue and Palette
sns.scatterplot(x="total_bill", y="tip", data=tips, hue="smoker", palette=hue_colors)
plt.show()
```

EXERCISE - Hue and scatter plots

```python
student_data = pd.read_csv('student-alcohol-consumption.csv')
student_data.head()
```

```python
# Create a scatter plot of absences vs. final grade
sns.scatterplot(x="absences", y="G3", data=student_data, hue="location")

# Show plot
plt.show()
```

```python
# Make "Rural" appear before "Urban" in the plot legend.
sns.scatterplot(x="absences", y="G3", data=student_data, hue="location", hue_order=["Rural","Urban"])

# Show plot
plt.show()
```

EXERCISE - Hue and count plots

```python
# Fill in the palette_colors dictionary to map the "Rural" location value to the color "green" and the "Urban" location value to the color "blue".
# Create a dictionary mapping subgroup values to colors
palette_colors = {"Rural": "green", "Urban": "blue"}

# Create a count plot with "school" on the x-axis using the student_data DataFrame.
# Create a count plot of school with location subgroups
sns.countplot(x="school", data=student_data, hue="location", palette=palette_colors)

# Display plot
plt.show()
```
